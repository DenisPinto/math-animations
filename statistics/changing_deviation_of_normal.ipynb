##### Exemplo 1: direto pra HTML5 Video

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from celluloid import Camera
from IPython.display import HTML

std = 3 # desvio padrão inicial
AnaPerf = 15. # média da primeira normal
MarPerf = 12.5 # média da segunda normal

fig, ax = plt.subplots(figsize=(9,6)) # criando o plot vazio
camera = Camera(fig)

# O celluloid funciona criando um objeto câmera que é usada para tirar "fotos" e depois
# as fotos são colocadas em sequência para criar uma animação no formato desejado

for j in range(10):
    plt.ylim((0, 0.2)) # se não colocar isso o gráfico vai se auto ajustar
    plt.xlim((-50, 50))
    
    Ana = np.random.normal(AnaPerf, std, size=(1000)) # criando as normais de tamnho 1000
    Mar = np.random.normal(MarPerf, std, size=(1000))

    if j == 0: # o primeiro plot tem labels para não ficar repetindo na hora do gráfico
        Ana_plot = sns.distplot(Ana, color='red', label='Ana') 
        Mar_plot = sns.distplot(Mar, color='blue', label='Mar')
        plt.legend()
    else:
        Ana_plot = sns.distplot(Ana, color='red')
        Mar_plot = sns.distplot(Mar, color='blue')
        
    ax.text(0.5, 1.01, "Desvio Padrão = "+str(std), transform=ax.transAxes) # fazendo o título
    camera.snap() # a foto da câmera
    
    std += 1 # incremento do desvio padrão

anim = camera.animate() # sequenciando as fotos
HTML(anim.to_html5_video()) # usando o HTML pro display
```
